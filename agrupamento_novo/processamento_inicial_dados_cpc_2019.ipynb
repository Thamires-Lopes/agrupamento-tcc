```python
import pandas as pd
from unidecode import unidecode
```

```python
excel_file_path_cpc = 'dados/resultados_cpc_2019.xlsx'
cpc = pd.read_excel(excel_file_path_cpc)
```

# CPC

### Funções

```python
def remove_acentos(texto):
    return unidecode(str(texto))
```

```python
def minusculo(texto):
    texto = texto.strip()
    texto = texto.replace('*', '')
    texto = texto.replace(' ', '_')
    return str(texto).lower()
```

```python
def is_inteiro(valor):
    try:
        int_valor = int(valor)
        return True
    except (ValueError, TypeError):
        return False
```

```python
def converter_para_int(valor):
    try:
        return int(valor)
    except ValueError:
        return 0
```

### Análise e limpeza dos dados

```python
cpc
```

#### Ajustando nomes das colunas e linhas

```python
colunas = cpc.columns
for coluna in colunas:
    print(coluna)
```

```python
cpc.columns = [minusculo(remove_acentos(col)) for col in cpc.columns]
```

```python
colunas_categoricas = cpc.select_dtypes(include=['object']).columns
for coluna in colunas_categoricas:
    cpc[coluna] = cpc[coluna].apply(remove_acentos)
    cpc[coluna] = cpc[coluna].apply(minusculo)
```

```python
colunas = cpc.columns
for coluna in colunas:
    print(coluna)
```
