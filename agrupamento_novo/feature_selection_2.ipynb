```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('dados_processados/cursos-pos-processamento.csv',sep=';')
```

```python
df_scaled = df.drop(['no_curso','co_curso'],axis=1)
```

```python
# Suponha que 'df' seja o seu DataFrame com o dataset carregado

# Configurações
mandatory_features = [
    "no_cine_area_geral_agricultura,_silvicultura,_pesca_e_veterinaria",
    "no_cine_area_geral_artes_e_humanidades",
    "no_cine_area_geral_ciencias_naturais,_matematica_e_estatistica",
    "no_cine_area_geral_ciencias_sociais,_comunicacao_e_informacao",
    "no_cine_area_geral_computacao_e_tecnologias_da_informacao_e_comunicacao_(tic)",
    "no_cine_area_geral_educacao",
    "no_cine_area_geral_engenharia,_producao_e_construcao",
    "no_cine_area_geral_negocios,_administracao_e_direito",
    "no_cine_area_geral_saude_e_bem-estar",
    "no_cine_area_geral_servicos"
]
n_clusters_range = range(2, 20)  # Por exemplo, de 2 a 10 clusters
max_features = 100  # Número máximo de features (incluindo as obrigatórias)
min_features = 15  # Número mínimo de features a serem usadas

# Features restantes para testar, excluindo as obrigatórias
candidate_features = [f for f in df_scaled.columns if f not in mandatory_features]

# Dicionário para armazenar o melhor score e o conjunto de features para cada k
best_results = {}

# Loop externo para cada número de clusters
for n_clusters in n_clusters_range:
    # Inicializa as variáveis para armazenar o melhor conjunto de features e scores
    best_features_pre_min = mandatory_features[:]  # Inicial com as features obrigatórias
    best_score_pre_min = -1
    best_features_post_min = mandatory_features[:]
    best_score_post_min = -1

    # Variável temporária para armazenar as features durante a construção do conjunto
    current_features = mandatory_features[:]

    # Loop interno para adicionar features uma por uma
    for feature in candidate_features:
        # Limita o número de features no conjunto atual
        if len(current_features) >= max_features:
            break

        # Adiciona a nova feature ao conjunto atual
        current_features.append(feature)

        # Executa o K-Means e calcula o índice de silhueta para o conjunto atual
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        clusters = kmeans.fit_predict(df_scaled[current_features])
        score = silhouette_score(df_scaled[current_features], clusters)

        # Armazena o melhor conjunto de features e score antes e depois do mínimo de features
        if len(current_features) < min_features:
            # Antes de atingir o mínimo de features
            if score > best_score_pre_min:
                best_score_pre_min = score
                best_features_pre_min = current_features[:]
        else:
            # Após atingir o mínimo de features
            if score > best_score_post_min:
                best_score_post_min = score
                best_features_post_min = current_features[:]

    # Armazena o melhor score e conjunto de features antes e depois do mínimo de features para o número atual de clusters
    best_results[n_clusters] = {
        "best_score_pre_min": best_score_pre_min,
        "best_features_pre_min": best_features_pre_min,
        "best_score_post_min": best_score_post_min,
        "best_features_post_min": best_features_post_min
    }

# Resultados finais
for k, result in best_results.items():
    print(f"Para k={k}:")
    print("Melhor conjunto de features antes do mínimo:", result["best_features_pre_min"])
    print("Melhor índice de silhueta antes do mínimo:", result["best_score_pre_min"])
    print("Melhor conjunto de features após o mínimo:", result["best_features_post_min"])
    print("Melhor índice de silhueta após o mínimo:", result["best_score_post_min"])
    print("---------")
```
