```python
import pandas as pd
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
import matplotlib.pyplot as plt
import warnings
```

```python
pd.set_option('display.max_columns',None)
warnings.filterwarnings("ignore")
```

```python
X = pd.read_csv('../dados_processados/cursos-ufrpe-pos-processamento.csv',sep=';')
```

```python
X.shape
```

```python
X.columns
```

```python
X_to_train = X.drop(['no_curso','co_curso'],axis=1)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=100)
kmeans.fit(X_to_train)
```

```python
visualizer = SilhouetteVisualizer(kmeans)
visualizer.fit(X_to_train)
visualizer.show()
```

```python
X_to_train['label'] = kmeans.labels_
```

```python
X_to_train['no_curso'] = X['no_curso']
```

```python
X_to_train
```

```python
X_to_train.label.value_counts()
```

```python
for i in range(0, 2):
    cursos_cluster_i = list(X_to_train.query('label == @i')['no_curso'])
    print('Cluster', i+1, '-', len(cursos_cluster_i), 'cursos')
    
    for curso in cursos_cluster_i:
        print(curso)
    
    print()
```
